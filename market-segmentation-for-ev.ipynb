```python

import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

# Dataset 1 ::: Statewise Charging Station Sanctioned ...It contains number of charging stations sanctioned by each government in Inida.


```python
dataset1= pd.read_csv("datasets/Statewise Charging Stations Sanctioned.csv")
```

```python
dataset1.head()
```

```python
dataset1.info()
```

```python
dataset1.shape
```

# Plotting a bar Graph using the above data

```python
#REMOVING THE TOTAL FROM THE DATASET
dataset1 = dataset1[dataset1['State'] != 'Total']
#AND NOW PLOTING 
plt.figure(figsize=(12, 6))
plt.bar(dataset1['State'], dataset1['Sanctioned State Wise Charging Stations'], color='#53c294')
plt.xlabel('States')
plt.ylabel('Number of Charging Stations')
plt.title('Number of Charging Stations Sanctioned State-wise')
plt.xticks(rotation=90)
plt.show()
```

Andhra Pradhesh , Madhaya Pradesh , Tamil Nadu , Maharashtra, Uttar Pradesh, and Gujarat have the highest number of sanctioned charging stations, with over 200 stations each.

```python
# Plotting pie chart for the distribution
plt.figure(figsize=(10, 10))
plt.pie(dataset1['Sanctioned State Wise Charging Stations'], labels=dataset1['State'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Charging Stations State-wise')
plt.show()

# Display total number of charging stations
totalChargingStations = sum(dataset1['Sanctioned State Wise Charging Stations'])
print(f'Total Charging Stations: {totalChargingStations}')
```


The distribution of charging stations is uneven across states, with a few states having a large number of stations and many states having no stations at all. This could lead to challenges for electric vehicle adoption in some parts of the country.

# Dataset3::: State wise operational Public EV Charging Stations (PCS) .....The Charging Stations which are built and are in operation.

Link for the Dataset:: https://pib.gov.in/PressReleasePage.aspx?PRID=1910392


```python
dataset3=pd.read_csv("datasets/Statewise Number Public Operational EVs .csv")
```

```python
dataset3.head()
```

```python
dataset3.info()
```

```python
dataset3.shape
```

```python
#REMOVING THE TOTAL FROM THE DATASET
dataset3 = dataset3[dataset3['State Name'] != 'Total']
#AND NOW PLOTING 
plt.figure(figsize=(12, 6))
plt.bar(dataset3['State Name'], dataset3['No. of Operational PCS'], color='#53c294')
plt.xlabel('States')
plt.ylabel('Number of Charging Stations')
plt.title('No. of Operational PCS')
plt.xticks(rotation=90)
plt.show()	

# Display total number of charging stations
totalChargingStations = sum(dataset3['No. of Operational PCS'])
print(f'Total Charging Stations: {totalChargingStations}')
```

##### Considering that the number of Sanctioned EV Charging Stations in the Dataset are 2877 and the total number of operational EV stations are over 6500.... there is a significant increase in  adoption of EV  in India. Delhi's lead in operational stations highlights urban areas' role in driving adoption, while states like Maharashtra, Karnataka, and Tamil Nadu follow suit, supporting extensive network expansion nationwide. WE can clearly say that Indian EV market is in rise as the number EV stations which are  being built are increasingly improving every  year.
